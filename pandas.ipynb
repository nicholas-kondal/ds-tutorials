# 9/11 - Cleaning Data, Casting Datatypes and Handling Missing Values

## People df

```python
import pandas as pd
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
df = pd.DataFrame(people)
df
```

```python
df.dropna() # drop all rows with None or NaN
# default values: df.dropna(axis='index', how='any')
# axis = 'columns' - drop columns, 'index' - drop rows
# how = 'any' - drop any rows/columns with missing values, 'all' - drop row/column only if it contains all missing values
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email']) # subset - only check some columns (how argument doesn't really do anything now, both last and email have to be missing for row to be dropped)
# add inplace=True to change in place
```

## Replacing data

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
df.isna() # returns mask (True = treated as N/A value)
```

```python
df.fillna(0) # fill N/A values with something
df.fillna('MISSING')
# add inplace to make change permanent
```

## Casting

```python
df.dtypes # object means column data type is String (or something non-numeric)
```

```python
df['age'].mean() # can't convert Strings to ints
```

```python
type(np.nan) # need to convert N/A values to floats
```

```python
df['age'] = df['age'].astype(float) # could do this or convert N/A values to e.g. 0 using fillna(), then cast to int instead
df.dtypes
# can also convert whole df using df.astype(TYPE)
```

```python
df['age'].mean()
```

# 10/11 - Dates and Time Series Data

```python
# ethereum cryptocurrency data broken down in hourly segments
import pandas as pd
df = pd.read_csv('pd11.csv')
df.head() # Dates are not DateTime objects
```

```python
df.loc[0, 'Date'].day_name() # reading in as string, not DateTime
```

```python
df['Date'] = pd.to_datetime(df['Date']) # will result in error since pandas doesn't know how to convert date automatically in its current format, need to pass in format string as argument to "format"
```

```python
# convert to date as you're loading data
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p') # second argument is the format string for the specific date format in this data - https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
df = pd.read_csv('pd11.csv', parse_dates=['Date'], date_parser=d_parser)
df.head()
```

```python
df.loc[0, 'Date'].day_name()
```

```python
df['Date'].dt.day_name() # run DateTime method on entire series using dt
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
df
```

```python
# latest and earliest date data
df['Date'].min()
df['Date'].max()
```

```python
# get difference between dates
df['Date'].max() - df['Date'].min()
```

```python
# see data within a date range
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
filt = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[filt]
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df['2019'] # same as using filter
df['2020-01':'2020-02']
```

```python
highs = df['High'].resample('D').max() # resample lets you see data by day (D), week (W) etc - https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
highs['2020-01-01']
```
