# 4/11 - Filtering

## People df

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
people_df = pd.DataFrame(people)
```

```python
# each condition returns True/False series object i.e. a mask ; |=or, &=and, ~=not
filt = (people_df['last'] == 'Schafer') | (people_df['first'] == 'John')
# people_df[~filt] # gives same as below
people_df.loc[~filt, 'email'] # can also get specific columns using loc
```

## Dev data df

```python
df = pd.read_csv('so_data/survey_results_public.csv', index_col='Respondent')
schema_df = pd.read_csv('so_data/survey_results_schema.csv', index_col='Column')
df.head()
```

```python
filt = df['LanguageWorkedWith'].str.contains('Python', na=False) # set what to do with NaN values (False = ignore NaN)
filt
```

```python
df.loc[filt, 'LanguageWorkedWith']
```

```python
high_salary = df['ConvertedComp'] > 70000
df.loc[high_salary] # df[high_salary] returns same thing
```

```python
df.loc[high_salary, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = df['Country'].isin(countries)
df.loc[filt, 'Country']
```
